By using colaborative filtering

```python
item_matrix = price_recommendation.pivot_table(index='CustomerID', columns='StockCode', values='Quantity',aggfunc='sum')
item_matrix.head()
```

```python
item_matrix = item_matrix.applymap(lambda x: 1 if x > 0 else 0)
item_matrix.head()
```

```python
item_matrix.shape
```

1. By Creating User-to-User Similarity Matrix

```python
from sklearn.metrics.pairwise import cosine_similarity

user_user_sim_matrix = pd.DataFrame(cosine_similarity(item_matrix))
user_user_sim_matrix
```

```python
user_user_sim_matrix.shape
```

```python
user_user_sim_matrix.columns = item_matrix.index

user_user_sim_matrix['CustomerID'] = item_matrix.index

user_user_sim_matrix = user_user_sim_matrix.set_index('CustomerID')
user_user_sim_matrix.head()
```

```python
user_user_sim_matrix.loc[12414].sort_values(ascending=False)
```

```python
items_bought_by_12414 = set(item_matrix.loc[12414].iloc[item_matrix.loc[12414].to_numpy().nonzero()].index)
items_bought_by_12414
```

```python
items_bought_by_12399 = set(item_matrix.loc[12399.0].iloc[item_matrix.loc[12399.0].to_numpy().nonzero()].index)
items_bought_by_12399
```

```python
items_to_recommend_to_12414 = items_bought_by_12414 - items_bought_by_12399
items_to_recommend_to_12414
```

```python
price_recommendation.loc[price_recommendation['StockCode'].isin(items_to_recommend_to_12414), ['StockCode', 'Description']].drop_duplicates().set_index('StockCode')
```

```python
most_similar_user = user_user_sim_matrix.loc[12414].sort_values(ascending=False).reset_index().iloc[1, 0]
most_similar_user
```

```python
def get_items_to_recommend_cust(cust_a): 
  most_similar_user = user_user_sim_matrix.loc[cust_a].sort_values(ascending=False).reset_index().iloc[1, 0]
  items_bought_by_cust_a = set(item_matrix.loc[cust_a].iloc[item_matrix.loc[cust_a].to_numpy().nonzero()].index)
  items_bought_by_cust_b = set(item_matrix.loc[most_similar_user].iloc[item_matrix.loc[most_similar_user].to_numpy().nonzero()].index)
  items_to_recommend_to_a = items_bought_by_cust_b - items_bought_by_cust_a
  items_description = price_recommendation.loc[price_recommendation['StockCode'].isin(items_to_recommend_to_a), ['StockCode', 'Description']].drop_duplicates().set_index('StockCode')
  return items_description
```

2. By Creating Ite to Item similarity matrix

```python
item_item_sim_matrix =pd.DataFrame(cosine_similarity(item_matrix.T))
item_item_sim_matrix.head()
```

```python
item_item_sim_matrix.shape
```

```python
item_item_sim_matrix.columns = item_matrix.T.index

item_item_sim_matrix['StockCode'] = item_matrix.T.index
item_item_sim_matrix = item_item_sim_matrix.set_index('StockCode')
item_item_sim_matrix.head()
```

```python
def get_top_similar_items(item):
  top_10_similar_items = list(item_item_sim_matrix.loc[item].sort_values(ascending=False).iloc[:10].index)
  top_10 = price_recommendation.loc[price_recommendation['StockCode'].isin(top_10_similar_items), ['StockCode', 'Description']].drop_duplicates().set_index('StockCode').loc[top_10_similar_items]
  return top_10
```
