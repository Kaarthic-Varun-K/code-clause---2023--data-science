```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
market_basket = pd.read_csv("market basket analysis.csv",encoding = 'latin1')
market_basket
```

```python
market_basket.head()
```

```python
market_basket.info()
```

Data cleansing

```python
market_basket['Description'] = market_basket['Description'].str.strip()
market_basket.dropna(axis=0 ,subset = ['InvoiceNo'] , inplace=True)
market_basket['InvoiceNo' ] = market_basket['InvoiceNo'].astype('str')
market_basket = market_basket[~ market_basket['InvoiceNo'].str.contains('C')]
```

```python
market_basket.shape
```

```python
market_basket.head()
```

```python
market_basket.describe()
```

```python
market_basket['Country'].value_counts()
```

```python
market_basket.shape
```

```python
basket = (market_basket[market_basket['Country'] =="EIRE"]     
          .groupby(['InvoiceNo', 'Description'])['Quantity'] 
          .sum().unstack().reset_index().fillna(0) 
          .set_index('InvoiceNo')) 
```

```python
basket.head()
```

convert positive value as 1 and negative value as 0

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
basket_sets = basket.applymap(encode_units)
```

```python
basket.shape
```

```python
basket_sets.head()
```

 Frequent item sets that have a support of at least 7%

```python
frequent_itemsets = apriori(basket_sets, min_support=0.07, use_colnames=True)
```

Generate the rules with their corresponding support, confidence and lift.

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
```

```python
rules.head()
```

```python
rules[(rules['lift'] >= 6) &
       (rules['confidence'] >= 0.7)]
```
